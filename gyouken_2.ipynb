```python
train_data = pd.read_csv("data_mart_171021.csv", encoding="shift-jis", header=0)
train_data = train_data.loc[(train_data.氏名id == 18)]
y_each = train_data.iloc[:,51]
X_each = train_data.iloc[:,[4, 6, 9]]

y_each = y_each.where(y_each < 1, 1)
X_train, X_test, y_train, y_test = train_test_split(X_each, y_each, test_size=0.3, random_state = 130)

sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)

lr = LogisticRegression(max_iter=30000)
lr.fit(X_train_std, y_train)

print (lr.score(X_test_std, y_test))

tmp = lr.predict_proba(X_test)[:,1]

print("LogisticRegression AUC =", roc_auc_score(y_test, tmp))
```
